```python
class employee:
    def _init_ (self):
        self.__salary=50000


    def updatedsalary(self,salary):
        if salary>0:
            self.__salary=salary
            return "updatedsalary"
        else:
            return "Invalid salary"
    def get_salary(self):
        return self.__salary

e1=employee()
e1.updatedsalary(10000)

print(e1.get_salary())
```
