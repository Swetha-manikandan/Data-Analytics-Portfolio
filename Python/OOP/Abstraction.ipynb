```python
class Employee():
    def work(self):
        print("Working")
class dev(Employee):
    def work(self):
        return "Developing"
m=dev()
print(m.work())   
class Hr(Employee):
    def work(self):
        return "Recruiting"
b=Hr()
print(b.work())

```

```python
from abc import ABC,abstractmethod
class employee(ABC):
    @abstractmethod
    def work(self):
        pass
```

```python
class developer(employee):
    def work(self):
        pass
```

```python
class HR(employee):
    def work(self):
        return "Recruit"
```

```python
class DA(employee):
    def work(self):
        return "Analyze"
```

```python
da1=DA()
print(da1.work())
```

```python
hr=HR()
print(hr.work())
```

```python
dev=developer()
print(dev.work())
```
